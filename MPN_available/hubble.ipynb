```python
import csv
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException, StaleElementReferenceException
import time

df = pd.read_csv(r'D:\raptor supplies\18DEC\hubell scrap.csv')

# Function to execute JavaScript and extract text
def execute_js_and_extract_text(driver, js_selector):
    try:
        return driver.execute_script(js_selector)
    except:
        return None

# Set up the webdriver
s = Service('D:\DATA SCIENCE\webscraping\selenium\chromedriver-win64 (1)\chromedriver-win64\chromedriver.exe')
driver = webdriver.Chrome(service=s)

# Navigate to the website
driver.get('https://www.hubbell.com/hubbell/en/search2?text=&CSRFToken=260fbd68-0315-40ae-bb80-08cfd3aef1ff')
time.sleep(5)

# Open CSV file for writing
with open('hubell_final2.csv', 'a', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['mpn_match','MPN','Title', 'Images', 'category', 'url', 'description', 'status']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    # Write headers to the CSV file
    writer.writeheader()

    for i in df['mpn'].iloc[5227:]:
        try:
            search = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, '/html/body/div[2]/div[2]/div[2]/div/div[1]/div[1]/div[2]/input'))
            )
            search.clear()
            search.send_keys(i)
            search.send_keys(Keys.ENTER)
            WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, '#search-results'))
            )
            time.sleep(2)
            mpn_path = '//div[@class="item-body item-body--section__product is-one-quarter column item-body--section-row"]//div[@class="item-body--attr"]'
            mpn_match = driver.find_element(By.XPATH, mpn_path).text
            print("i:", i)
            print("mpn_match:", mpn_match)
            if str(i).strip() == str(mpn_match).strip():
                # #if discountinue path is not found give ok
                discontinued = []
                dis_path = '//div[@class="search-item"][1] //div[@class="item-body--discontinued"]'

                try:
                    discontinue = driver.find_element(By.XPATH, dis_path).text
                    discontinued.append(discontinue)
                except NoSuchElementException:
                    discontinue = "OK"




                title_path = '#search-results > div:nth-child(2) > div > div.results.column.is-three-quarters > div > div:nth-child(1) > div > div.product-item.result-item.result-item--container > div.result-item.result-item--body > div.item-body.item-body--section__product.is-three-quarters.column.item-body--order-two > div.item-body--title > a.item-body--long-product-name'
                title_js_selector = f'return document.querySelector("{title_path}").textContent;'

                url_js_selector = f'return document.querySelector("{title_path}").getAttribute("href");'

                img_path = '.result-item--container .item-body--img-wrapper a div img'
                image_js_selector = f'return document.querySelector("{img_path}").getAttribute("src");'

                category_path1 = '.result-item--container .item-body--desc-wrapper__pc div:nth-child(2)'
                category_js_selector1 = f'return document.querySelector("{category_path1}").textContent;'

                category_path2 = '.result-item--container .item-body--desc-wrapper__pc div:nth-child(3)'
                category_js_selector2 = f'return document.querySelector("{category_path2}").textContent;'

                description_path = '#search-results > div:nth-child(2) > div > div.results.column.is-three-quarters > div > div > div > div.product-item.result-item.result-item--container > div.result-item.result-item--body > div.item-body.item-body--section__product.is-three-quarters.column.item-body--order-two > a'
                description_js_selector = f'return document.querySelector("{description_path}").textContent;'

                description = execute_js_and_extract_text(driver, description_js_selector)

                if description is None:
                    description = "Description not found"

                # Extract url using JavaScript
                url = execute_js_and_extract_text(driver, url_js_selector)
                if url is None:
                    url = "url not found"

                title = execute_js_and_extract_text(driver, title_js_selector)
                if title is None:
                    title = "Title not found"

                # Extract images
                images = execute_js_and_extract_text(driver, image_js_selector)
                if images is None:
                    images = 'Images not found'

                if mpn_match is None:
                    mpn_match = "mpn not found"

                # if discontinued is None:
                #     discontinued = "OK"
                
                

                # Extract categories
                categories1 = execute_js_and_extract_text(driver, category_js_selector1)
                categories1 = categories1.split(',') if categories1 else []

                categories2 = execute_js_and_extract_text(driver, category_js_selector2)
                categories2 = categories2.split(',') if categories2 else []

                category = [categories1, categories2]

                # Write the data to the CSV file
                writer.writerow({
                    'Title': title,
                    'MPN': i,
                    "mpn_match": mpn_match,
                    'Images': images,
                    'category': category,
                    'url': url,
                    'description': description,
                    'status':discontinued
                })
                # print('mpn_match',mpn_match)
                # print('mpn', i)
                print("Title:", title)
                print("Images:", images)
                print("category", category)
                print('url', url)
                print('description', description)
                print('status:',discontinued)
                print('________________')
            else:
                print('not found')

        except (TimeoutException,NoSuchElementException, StaleElementReferenceException):
            print(f"not found: {i}")
            driver.refresh()

driver.quit()
```
